```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style('darkgrid')
```

**Load Dataset**

```python
df=pd.read_csv('data.csv')
df=df.drop("Unnamed: 0",axis=1)
df.head()
```

**Data Cleaning**

```python
for x in df.columns:
    print(x)
```

```python
df.describe()
```

**Data Analysis**

**Top 5 popular artists:**

```python
top_5_artists=df.groupby("artist").count().sort_values(by='song_title',ascending=False)["song_title"][:5]
top_5_artists
```

```python
top_5_artists.plot.barh()
plt.show()
```

**Top 5 loudest tracks:**

```python
top_5_loud_tracks=df[["loudness",'song_title']].sort_values(by='loudness',ascending=False)[-5:]
top_5_loud_tracks
```

**negative dB implies more loud sound**

```python
plt.figure(figsize=(12,7))
sns.barplot(x='loudness',y='song_title',data=top_5_loud_tracks)
plt.title('Top 5 loudest tracks')
plt.show()
```

**Artisis with most danceability song:**

```python
top_5_artists_dance=df[["danceability","song_title","artist"]].sort_values(by='danceability',ascending=False)[:5]
top_5_artists_dance
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x='danceability',y='artist',data=top_5_artists_dance)
plt.title('Artists with most danceability songs')
plt.show()
```

**Top 10 instrumental tracks:**

```python
top_10_instrumental_tracks=df[['instrumentalness','song_title']].sort_values(by='instrumentalness',ascending=False)[:10]
top_10_instrumental_tracks
```

```python
plt.figure(figsize=(12,7))
sns.barplot(y='song_title',x='instrumentalness',data=top_10_instrumental_tracks)
plt.title('Top 10 instrumental tracks')
plt.show()
```

```python
plt.figure(figsize=(12,7))
plt.pie(x='instrumentalness',data=top_10_instrumental_tracks,autopct='%1.2f%%',labels=top_10_instrumental_tracks.song_title)
plt.title('instrumentalness of top 10 songs')
plt.show()
```

**Multiple feature plot:**

```python
df.columns
```

```python
interest_feature_cols=list(df.columns)
```

```python
print(interest_feature_cols)
```

```python
for feature_col in interest_feature_cols:
    pos_data=df[df['target']==1][feature_col]
    neg_data=df[df['target']==0][feature_col]
    
    plt.figure(figsize=(12,7))
    
    sns.distplot(pos_data,bins=30,label='positive data',color='green')
    sns.distplot(neg_data,bins=30,label='negative data',color='red')
    
    plt.legend(loc='upper right')
    plt.title(f"Positive and Negative Histogram PLot for {feature_col}")
    plt.show()
```
